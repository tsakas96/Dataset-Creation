```python
import cv2
import numpy as np
import numpy
import random

def findPNGFileNameCategoryAndPath(folder_src_path):
    file_names_categories_paths = set()
    for src_path, _, filenames in os.walk(folder_src_path):
        for filename in [f for f in filenames if f.endswith('.png')]:
            #category + name
            category = src_path.replace(folder_src_path + "\\", "")
            file_names_categories_paths.add((filename, category, src_path + "\\" + filename))
    return file_names_categories_paths

def create_directory(path):
    try:
        os.mkdir(path)
    except OSError:
        print ("Creation of the directory %s failed" % path)
    else:
        print ("Successfully created the directory %s " % path)
```

```python
src_path = 'C:\\Users\\arist\\Desktop\\Dataset2\\RenamedSketches'

file_names_categories_paths = findPNGFileNameCategoryAndPath(src_path)

width = 256
height = 256
dim = (width, height)
dst_path = 'C:\\Users\\arist\\Desktop\\Dataset2\\ResizedSketches\\'
create_directory(dst_path)
for name, category, path in file_names_categories_paths:
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)

    create_directory(dst_path + category)
    # resize image
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    cv2.imwrite(dst_path + category + "\\" + name, resized)
```

```python
scale_list = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1]
src_path = 'C:\\Users\\arist\\Desktop\\Dataset2\\Augmentations\\ResizedSketches'
file_names_categories_paths = findPNGFileNameCategoryAndPath(src_path)

dst_path = 'C:\\Users\\arist\\Desktop\\Dataset2\\Augmentations\\ScaleAndPaddedSketches\\'
create_directory(dst_path)
for name, category, path in file_names_categories_paths:
    img = cv2.imread(path)

    #########Scale image##################
    index_scale = random.randint(0, 10)
    scale = (scale_list[index_scale], scale_list[index_scale]) # define your scale
    scaled_img = cv2.resize(img, None, fx=scale[0], fy=scale[1]) # scale image

    h, w = img.shape[:2] # get h, w of image
    sh, sw = scaled_img.shape[:2] # get h, w of scaled image

    # randomize the position of photo
    center_y = random.randint(0,h-sh)
    center_x = random.randint(0,w-sw)
    padded_scaled = np.ones(img.shape, dtype=np.uint8)*255 # using img.shape to obtain #channels
    padded_scaled[center_y:center_y+sh, center_x:center_x+sw] = scaled_img
    ##########################################

    create_directory(dst_path + category)
    cv2.imwrite(dst_path + category + "\\" + name, padded_scaled)
```
