```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import matplotlib.ticker as ticker
```

```python
folder_path = "C:\\Users\\arist\\Desktop\\Results-Cluster\\GoogleNet\\Triplet Loss-CWI-HumanTest-Seen\\10-06-2021 02-34-58\\"
file_path = folder_path + "stats.txt"
epoch_list = []
loss_list = []
top1_list = []
top10_list = []
with open(file_path, "r") as filestream:
    for line in filestream:
        currentline = line.split(",")
        epoch_list.append(int(currentline[0]))
        loss_list.append(float(currentline[1]))
        top1_list.append(float(currentline[2])*100)
        top10_list.append(float(currentline[3])*100)
```

```python
f, ax = plt.subplots(1)
plt.title('Loss vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
ax.plot(epoch_list, loss_list)
ax.set_ylim(ymin=0, ymax = 50)
ax.set_xlim(xmin=0)
plt.show(f)
```

```python
fontP = FontProperties()
fontP.set_size('large')
f, ax = plt.subplots(1)
plt.title('Accuracy-k vs Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy-k')
p1, = ax.plot(epoch_list, top1_list, label='acc @1')
p2, = ax.plot(epoch_list, top10_list, label='acc @10')
ax.set_ylim(ymin=0)
ax.set_xlim(xmin=0)
plt.legend(handles=[p1, p2], title='Accuracy', bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP)
plt.show(f)
```
