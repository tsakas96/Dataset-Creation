```python
import yaml
import os
import os.path
from shutil import copyfile
import collections

# it creates a directory to the path
# eg. the path is 'C:\\Users\\arist\\Desktop\\icon'
# and it creates folder 'icon' into the path: 'C:\\Users\\arist\\Desktop\\'
def create_directory(path):
    try:
        os.mkdir(path)
    except OSError:
        print ("Creation of the directory %s failed" % path)
    else:
        print ("Successfully created the directory %s " % path)

# search through all subfolders of the folder_src_path and it copies
# the icon into the folder_dst_path. If store dublicates is true
# we can store photos which are found in different folders with the same
# name eg. if bell.svg is found in two directories then i will store
# bell.svg and bell1.svg
def copySVGsToCategoryFolder(folder_src_path, folder_dst_path, icon):
    store_duplicates = True
    count = 0
    for src_path, src_names, filenames in os.walk(folder_src_path):
        for filename in [f for f in filenames if f == icon]:
            icon_old_name = icon
            
            if count > 0:
                if store_duplicates:
                    icon = icon.replace('.svg',"")
                    icon = icon + str(count) + '.svg'

            count = count + 1
            src = src_path + '\\' + icon_old_name
            dst = folder_dst_path + '\\' + icon
            copyfile(src, dst)

# the path where categories.yml is stored
categories_path = 'C:\\Users\\arist\\Desktop\\fontawesome\\MyCategories.yml'
# open the file and store its contents into documents dictionary
with open(categories_path, 'r') as file:
    documents = yaml.safe_load(file)

# create a direcoty "icon" into 'C:\\Users\\arist\\Desktop
sketch_path = 'C:\\Users\\arist\\Desktop\\icon'
create_directory(sketch_path)

#path where the icons are stored (it has 3 subfolders)
folder_src_path = 'C:\\Users\\arist\\Desktop\\fontawesome\\svgs'

# it takes the categories in the documents dictionary and it creates 
# subfolders into "sketch" with the category as a name
# foreach icon which in the category, it searches through the folder_src_path
# and when it finds the icon (.svg) it copies it into the category folder
icons_list = []
for category, doc in documents.items():
    categories_path = sketch_path + '\\' + category 
    create_directory(categories_path)
    for icon in doc['icons']:
        if icon not in icons_list:
            icons_list.append(icon)
            icon = icon + '.svg'
            copySVGsToCategoryFolder(folder_src_path, categories_path, icon)

```

```python
dataset_list = []
for src_path, src_names, filenames in os.walk(folder_src_path):
    for filename in [f for f in filenames if f.endswith('.svg')]:
        dataset_list.append(filename.replace('.svg',""))

uncommon = list(set(dataset_list) - set(icons_list))
dublicates = [item for item, count in collections.Counter(dataset_list).items() if count > 1]
print("All the files in the 3 folders are: " + str(len(dataset_list)))
print("All the files i store in icons with no dublicates: " + str(len(icons_list)))
print("The dublicate files i store in icons: " + str(len(dublicates)))
print("The files that are uncommon in the two lists: " + str(len(uncommon)))


#for unc in uncommon:
 #   for src_path, src_names, filenames in os.walk(folder_src_path):
 #       for filename in [f for f in filenames if f.endswith(unc + '.svg')]:
  #          print(src_path + '\\' + unc + '.svg')
```
